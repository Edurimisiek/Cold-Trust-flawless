```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import cifar10
```

```python
def create_model(hidden_layers=[32, 64], units_per_layer=32,
                 loss='categorical_crossentropy', activation='relu',
                 optimizer='adam', dropout_rate=0.2, regularizer=None):
    model = tf.keras.Sequential([
        layers.Flatten(input_shape=(32, 32, 3)),
        # Add hidden layers with specified parameters
        *[layers.Dense(units, activation=activation, kernel_regularizer=regularizer)
          for units in hidden_layers],
        layers.Dense(10, activation='softmax')  # Output layer
    ])
    model.compile(loss=loss, optimizer=optimizer, metrics=['accuracy'])
    return model
# Define options for hyperparameters
hidden_layers_options = [[32], [32, 64], [64, 32, 64]]
units_per_layer_options = [16, 32, 64]
loss_options = ['categorical_crossentropy', 'sparse_categorical_crossentropy']
activation_options = ['relu', 'tanh']

# Iterate through combinations and train/evaluate models
for hidden_layers in hidden_layers_options:
    for units_per_layer in units_per_layer_options:
        for loss in loss_options:
            for activation in activation_options:
                model = create_model(hidden_layers, units_per_layer, loss, activation,
                                     dropout_rate=0.5, regularizer=tf.keras.regularizers.l2(0.01))
                # Train and evaluate the model
```
